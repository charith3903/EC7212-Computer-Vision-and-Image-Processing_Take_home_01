```python

from google.colab import files
from PIL import Image
import io
import numpy as np
import matplotlib.pyplot as plt

# Upload
uploaded = files.upload()

# Load
for filename in uploaded.keys():
    img = Image.open(io.BytesIO(uploaded[filename]))
    img = img.convert('RGB')  # ensure RGB
    img_array = np.array(img)
    break

# Show
plt.imshow(img_array)
plt.title("Original Color Image")
plt.axis('off')
plt.show()


def block_average(image, block_size):
    (h, w, c) = image.shape
    output = image.copy()
    for y in range(0, h, block_size):
        for x in range(0, w, block_size):
            block = image[y:y+block_size, x:x+block_size]
            avg_color = block.reshape(-1, c).mean(axis=0).astype(np.uint8)
            output[y:y+block_size, x:x+block_size] = avg_color
    return output

for block in [3, 5, 7]:
    reduced = block_average(img_array, block)
    plt.imshow(reduced)
    plt.title(f"{block}x{block} Block-wise Average")
    plt.axis('off')
    plt.show()
```
